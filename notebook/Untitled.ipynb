```python
with open('WPACV.yml') as fp:
    wpacv = yaml.safe_load(fp)
    
```

```python
amarna_names = []
for name in wpacv['Names']:
    try:
        for key, item in name.items():
            try:
                for form in item["UF"]:
                    amarna_names.append((form, key))
            except KeyError:
                pass
    except AttributeError:
        pass
```
